````python
import pandas as pd
from io import StringIO

def process_insurance_data(data: str) -> pd.DataFrame:
    # Read data into a DataFrame
    df = pd.read_csv(StringIO(data), header=None)
    
    # Define the column names based on the data provided
    df.columns = ["ID", "Policy_ID", "State", "Policy_Annual_Premium", "Renew_Offer_Type", 
                  "Sales_Channel", "Customer_Lifetime_Value", "Coverage", "Education", 
                  "EmploymentStatus", "Gender", "Income", "Location_Code", "Marital_Status", 
                  "Monthly_Premium_Auto", "Months_Since_Last_Claim", "Months_Since_Policy_Inception",
                  "Number_of_Open_Complaints", "Number_of_Policies", "Policy_Type", 
                  "Policy", "Renew_Offer_Type", "Sales_Channel", "Total_Claim_Amount", 
                  "Vehicle_Class", "Vehicle_Size", "Effective_Date"]
    
    # Fill missing values in 'Vehicle_Size' and 'Vehicle_Class' with 'Unknown'
    df.fillna({'Vehicle_Size': 'Unknown', 'Vehicle_Class': 'Unknown'}, inplace=True)
    
    # Extract month from 'Effective_Date'
    df['Effective_Date'] = pd.to_datetime(df['Effective_Date'], format='%m/%d/%y', errors='coerce')
    df['Month'] = df['Effective_Date'].dt.month
    
    # Filter for the first quarter
    df_first_quarter = df[df['Month'].isin([1, 2, 3])]
    
    return df_first_quarter

data = """
4781,JU55710,Oregon,3746.751625,Yes,Extended,College,2/26/11,Employed,F,41479,Rural,Married,94,14.0,38,1.0,1,Personal Auto,Personal L1,Offer2,Call Center,19.575683,Two-Door Car,Large,nan
4782,QT39025,Arizona,4582.899414,No,Basic,Bachelor,1/18/11,Employed,F,66991,Urban,Married,115,1.0,71,0.0,1,Personal Auto,Personal L3,Offer2,Branch,399.617715,SUV,Medsize,nan
4783,WG15841,California,2810.464447,Yes,Basic,Bachelor,2/27/11,Employed,M,26790,Suburban,Married,71,4.0,41,0.0,1,Personal Auto,Personal L3,Offer2,Agent,340.8,Two-Door Car,Medsize,nan
4784,WW99844,Oregon,4944.088762,No,Basic,High School or Below,2/7/11,Employed,F,53607,Rural,Married,61,4.0,18,0.0,9,Personal Auto,Personal L2,Offer3,Web,93.891207,Four-Door Car,Medsize,nan
4785,KC75958,California,16468.22079,Yes,Basic,High School or Below,1/27/11,Unemployed,F,0,Suburban,Divorced,63,4.0,52,0.0,2,Personal Auto,Personal L1,Offer2,Agent,685.12694,Four-Door Car,Medsize,A
4786,PO26937,Oregon,8223.164916,No,Premium,Bachelor,1/18/11,Employed,M,95102,Rural,Married,101,6.0,73,0.0,8,Corporate Auto,Corporate L3,Offer3,Web,94.338814,Four-Door Car,Medsize,nan
4787,CW82151,California,8902.737595,No,Basic,Bachelor,1/5/11,Employed,M,46833,Rural,Married,112,6.0,3,0.0,9,Personal Auto,Personal L3,Offer1,Call Center,64.109663,Sports Car,Medsize,nan
4788,KW52549,Arizona,27789.69235,Yes,Extended,Bachelor,1/20/11,Employed,F,33806,Urban,Divorced,89,32.0,59,0.0,2,Personal Auto,Personal L3,Offer2,Web,395.729716,Four-Door Car,Medsize,A
4789,BV63942,,9002.409572,,Premium,High School or Below,2/7/11,Employed,M,76184,Suburban,Married,111,23.0,19,0.0,3,Corporate Auto,Corporate L3,Offer2,Agent,532.8,Two-Door Car,Small,nan
4790,HT48275,Oregon,5532.753182,No,Basic,Bachelor,1/19/11,Employed,M,91212,Urban,Married,69,9.0,92,0.0,6,Personal Auto,Personal L3,Offer2,Call Center,270.311239,Two-Door Car,Medsize,A
4791,XX12304,California,2924.976721,No,Basic,College,1/9/11,Employed,F,64459,Urban,Married,72,16.0,64,0.0,1,Personal Auto,Personal L1,Offer2,Branch,240.259479,Four-Door Car,Medsize,nan
4792,WL74893,Oregon,4529.112791,No,Basic,College,2/24/11,Unemployed,F,0,Urban,Married,62,22.0,57,0.0,9,Special Auto,Special L3,Offer3,Agent,271.164139,Four-Door Car,Medsize,nan
4793,LY97989,Oregon,2894.243869,No,Extended,Bachelor,2/17/11,Unemployed,F,0,Suburban,Single,85,3.0,53,0.0,1,Personal Auto,Personal L3,Offer3,Agent,408.0,Four-Door Car,Medsize,nan
4794,NN40477,Arizona,4280.747008,No,Basic,Bachelor,2/9/11,Employed,F,52679,Suburban,Divorced,110,25.0,21,2.0,1,Personal Auto,Personal L2,Offer2,Branch,528.0,Sports Car,Medsize,nan
4795,QP80741,Washington,6048.574243,No,Basic,College,1/1/11,Unemployed,F,0,Suburban,Married,61,17.0,71,3.0,2,Corporate Auto,Corporate L1,Offer3,Agent,292.8,Four-Door Car,Medsize,A
4796,VF21395,Oregon,2652.695343,No,Basic,Bachelor,1/1/11,Employed,M,82655,Suburban,Married,66,8.0,93,0.0,1,Personal Auto,Personal L2,Offer2,Web,316.8,Two-Door Car,Medsize,nan
4797,PU73561,Arizona,4196.429642,No,Basic,Bachelor,2/14/11,Employed,F,24641,Suburban,Married,105,12.0,82,0.0,1,Personal Auto,Personal L2,Offer2,Agent,504.0,SUV,Medsize,A
4798,OQ88395,,4510.542608,,Basic,College,2/4/11,Employed,M,84300,Rural,Married,111,13.0,0,0.0,1,Corporate Auto,Corporate L3,Offer2,Branch,202.22495,,,A
4799,IN64985,Nevada,6140.242989,No,Extended,College,1/26/11,Employed,F,34112,Suburban,Married,77,32.0,68,0.0,6,Personal Auto,Personal L3,Offer2,Branch,369.6,Four-Door Car,Small,nan
4800,GY99465,Oregon,6869.185344,No,Basic,College,2/19/11,Unemployed,F,0,Suburban,Single,69,18.0,71,2.0,2,Corporate Auto,Corporate L2,Offer2,Branch,331.2,Four-Door Car,Medsize,A
4801,HW56522,Arizona,19185.0725,No,Extended,Bachelor,2/19/11,Employed,M,35322,Rural,Married,244,4.0,33,0.0,5,Personal Auto,Personal L3,Offer1,Agent,76.826503,Luxury Car,Small,A
4802,MV89967,Oregon,23142.1509,No,Premium,College,1/8/11,Unemployed,F,0,Suburban,Single,105,16.0,16,0.0,2,Personal Auto,Personal L2,Offer1,Agent,564.947278,Four-Door Car,Medsize,nan
4803,JQ35849,California,5360.557066,No,Basic,High School or Below,1/21/11,Employed,M,24254,Suburban,Married,68,11.0,84,0.0,5,Personal Auto,Personal L3,Offer4,Agent,326.4,Four-Door Car,Medsize,nan
4804,OU50429,California,7524.735746,No,Basic,Bachelor,2/4/11,Employed,M,36580,Suburban,Married,64,20.0,76,1.0,2,Personal Auto,Personal L3,Offer2,Agent,460.8,Four-Door Car,Small,nan
4805,CQ75652,Oregon,8341.623748,No,Premium,College,1/8/11,Unemployed,M,0,Suburban,Divorced,118,15.0,77,0.0,4,Personal Auto,Personal L2,Offer1,Agent,566.4,Four-Door Car,Medsize,nan
4806,ZB84321,Washington,19690.44179,No,Basic,College,1/30/11,Employed,F,36032,Suburban,Divorced,70,11.0,39,0.0,2,Personal Auto,Personal L3,Offer3,Web,336.0,Four-Door Car,Small,A
4807,CC91503,,10169.36981,,Extended,High School or Below,1/3/11,Unemployed,F,0,Suburban,Married,135,32.0,61,0.0,8,Personal Auto,Personal L3,Offer1,Agent,648.0,Sports Car,Medsize,nan
4808,XS89049,,2822.591481,,Basic,College,2/9/11,Disabled,F,25531,Suburban,Married,74,27.0,52,0.0,1,Personal Auto,Personal L3,Offer1,Agent,355.2,Two-Door Car,Small,nan
4809,PG26401,Washington,9191.672825,No,Basic,High School or Below,1/24/11,Employed,M,77603,Suburban,Married,114,10.0,83,0.0,8,Personal Auto,Personal L1,Offer2,Call Center,547.2,SUV,Medsize,nan
4810,WV17090,California,25585.72783,No,Basic,Bachelor,2/1/11,Unemployed,M,0,Suburban,Married,116,0.0,45,0.0,2,Corporate Auto,Corporate L1,Offer2,Call Center,830.623064,SUV,Medsize,A
4811,HA56608,California,13727.79972,Yes,Extended,College,1/8/11,Unemployed,M,0,Suburban,Single,126,12.0,4,0.0,2,Special Auto,Special L3,Offer1,Agent,604.8,SUV,Medsize,A
4812,HP20623,California,5227.570224,No,Basic,High School or Below,1/13/11,Employed,F,70988,Suburban,Married,65,3.0,82,0.0,7,Personal Auto,Personal L1,Offer3,Web,312.0,Four-Door Car,Medsize,nan
4813,BB48165,California,11095.97445,No,Basic,Bachelor,1/3/11,Unemployed,F,0,Suburban,Single,101,10.0,48,0.0,2,Personal Auto,Personal L3,Offer1,Agent,727.2,SUV,Medsize,A
4814,BV80674,Oregon,4340.858797,No,Extended,Bachelor,2/6/11,Unemployed,F,0,Suburban,Single,124,6.0,34,0.0,1,Personal Auto,Personal L2,Offer3,Call Center,595.2,SUV,Medsize,nan
4815,ZR32858,California,4315.788632,No,Basic,Bachelor,2/25/11,Employed,F,41252,Urban,Married,109,25.0,31,0.0,1,Personal Auto,Personal L3,Offer2,Branch,514.059402,SUV,Medsize,nan
4816,QD17769,Oregon,2552.038766,No,Basic,High School or Below,2/12/11,Employed,M,75740,Rural,Married,64,13.0,11,0.0,1,Corporate Auto,Corporate L3,Offer3,Agent,128.516525,Two-Door Car,Medsize,nan
4817,UH62254,,21423.63723,,Basic,Bachelor,1/3/11,Unemployed,M,0,Suburban,Married,65,9.0,31,0.0,2,Personal Auto,Personal L2,Offer2,Agent,312.0,Two-Door Car,Medsize,nan
4818,VC34797,California,2435.535995,No,Basic,Bachelor,1/27/11,Employed,F,40921,Rural,Married,62,0.0,50,2.0,1,Corporate Auto,Corporate L2,Offer3,Agent,10.560419,Four-Door Car,Medsize,nan
4819,TM28833,Oregon,2276.392936,No,Basic,College,1/18/11,Unemployed,M,0,Suburban,Single,69,,13,,1,Personal Auto,Personal L3,Offer4,Call Center,624.980751,Two-Door Car,Small,A
4820,GM62071,Washington,21435.87782,No,Basic,High School or Below,1/7/11,Employed,F,20668,Suburban,Married,188,34.0,54,3.0,2,Personal Auto,Personal L3,Offer2,Call Center,902.4,,,A
4821,UF91715,Nevada,7664.80924,No,Premium,High School or Below,1/22/11,Unemployed,M,0,Suburban,Single,107,14.0,90,0.0,3,Personal Auto,Personal L3,Offer1,Agent,770.4,Four-Door Car,Medsize,nan
4822,JV12632,Washington,2858.992036,No,Extended,Master,1/20/11,Unemployed,F,0,Urban,Married,83,25.0,91,4.0,1,Corporate Auto,Corporate L2,Offer4,Branch,355.385287,Four-Door Car,Medsize,nan
4823,UD31660,California,4061.590219,No,Premium,Bachelor,1/13/11,Employed,M,70129,Rural,Married,103,24.0,57,1.0,1,Personal Auto,Personal L2,Offer1,Branch,130.495798,Four-Door Car,Medsize,A
4824,BU81591,California,10153.88899,No,Extended,College,2/15/11,Unemployed,M,0,Suburban,Married,90,33.0,75,0.0,2,Corporate Auto,Corporate L1,Offer1,Call Center,432.0,Four-Door Car,Large,A
4825,PS23376,California,5671.09467,No,Basic,High School or Below,2/26/11,Employed,F,69668,Rural,Married,70,34.0,33,0.0,4,Corporate Auto,Corporate L2,Offer1,Branch,64.855035,Two-Door Car,Large,nan
4826,CC43119,California,2921.490461,No,Basic,Bachelor,1/19/11,Employed,F,62018,Rural,Married,73,2.0,59,0.0,1,Corporate Auto,Corporate L1,Offer4,Web,90.094835,Two-Door Car,Medsize,nan
4827,DI30528,California,2725.356404,Yes,Basic,Master,1/3/11,Employed,F,36650,Rural,Married,69,2.0,70,1.0,1,Personal Auto,Personal L3,Offer2,Web,56.60333,Four-Door Car,Large,nan
4828,PU57907,California,3014.498658,Yes,Basic,College,1/20/11,Unemployed,F,0,Suburban,Single,83,10.0,22,1.0,1,Personal Auto,Personal L3,Offer1,Branch,547.2,SUV,Small,nan
4829,ZG49224,California,17113.09014,Yes,Basic,High School or Below,1/16/11,Unemployed,F,0,Urban,Married,65,8.0,51,0.0,1,Personal Auto,Personal L2,Offer1,Web,239.381218,Two-Door Car,Small,A
4830,BG89267,California,11424.22048,No,Basic,Bachelor,2/20/11,Employed,F,70129,Urban,Married,103,3.0,36,0.0,1,Personal Auto,Personal L3,Offer1,Branch,556.8,Four-Door Car,Medsize,nan
4831,DL75666,Oregon,4714.021544,Yes,Basic,College,2/2/11,Employed,M,36098,Rural,Single,84,27.0,19,0.0,1,Personal Auto,Personal L3,Offer1,Branch,302.4,Four-Door Car,Small,A
4832,YX39771,California,26072.73855,No,Basic,College,2/27/11,Employed,M,26666,Suburban,Married,84,22.0,33,0.0,1,Personal Auto,Personal L3,Offer2,Agent,192.0,Two-Door Car,Small,nan
4833,LO61490,California,3832.622858,No,Basic,Master,1/14/11,Employed,F,68572,Rural,Married,70,8.0,4,0.0,2,Personal Auto,Personal L3,Offer2,Branch,368.923548,Four-Door Car,Medsize,A
4834,VG35884,Oregon,3821.205298,No,Premium,Bachelor,2/5/11,Employed,M,80954,Rural,Married,100,2.0,23,0.0,2,Personal Auto,Personal L1,Offer4,Agent,636.0,Four-Door Car,Medsize,A
"""

# Process the data
df_first_quarter = process_insurance_data(data)

# Display the DataFrame
import ace_tools as tools; tools.display_dataframe_to_user(name="First Quarter Insurance Data", dataframe=df_first_quarter)

df_first_quarter.head()
````

````python
import pandas as pd
from io import StringIO

def process_insurance_data(data):
    # Read the data into a DataFrame
    df = pd.read_csv(StringIO(data), header=None)
    
    # Define the column names based on the data provided
    df.columns = ["ID", "Policy_ID", "State", "Policy_Annual_Premium", "Renew_Offer_Type",
                  "Sales_Channel", "Customer_Lifetime_Value", "Coverage", "Education",
                  "EmploymentStatus", "Gender", "Income", "Location_Code", "Marital_Status",
                  "Monthly_Premium_Auto", "Months_Since_Last_Claim", "Months_Since_Policy_Inception",
                  "Number_of_Open_Complaints", "Number_of_Policies", "Policy_Type",
                  "Policy", "Sales_Channel_2", "Total_Claim_Amount", "Vehicle_Class", "Vehicle_Size",
                  "Effective_Date", "Unknown"]
    
    # Drop the 'Unknown' column as it seems to be an extra column
    df.drop(columns=['Unknown'], inplace=True)

    # Fill missing values in 'Vehicle_Size' and 'Vehicle_Class' with 'Unknown'
    df.fillna({'Vehicle_Size': 'Unknown', 'Vehicle_Class': 'Unknown'}, inplace=True)
    
    # Extract month from 'Effective_Date'
    df['Effective_Date'] = pd.to_datetime(df['Effective_Date'], format='%m/%d/%y', errors='coerce')
    df['Month'] = df['Effective_Date'].dt.month
    
    # Filter for the first quarter
    df_first_quarter = df[df['Month'].isin([1, 2, 3])]
    
    return df_first_quarter

data = """
4781,JU55710,Oregon,3746.751625,Yes,Extended,College,2/26/11,Employed,F,41479,Rural,Married,94,14.0,38,1.0,1,Personal Auto,Personal L1,Offer2,Call Center,19.575683,Two-Door Car,Large,nan
4782,QT39025,Arizona,4582.899414,No,Basic,Bachelor,1/18/11,Employed,F,66991,Urban,Married,115,1.0,71,0.0,1,Personal Auto,Personal L3,Offer2,Branch,399.617715,SUV,Medsize,nan
4783,FF45793,California,6193.817509,Yes,Extended,Bachelor,2/25/11,Unemployed,F,0,Suburban,Divorced,66,14.0,11,0.0,5,Personal Auto,Personal L3,Offer1,Agent,153.494597,SUV,Small,A
4784,BL84792,Washington,8536.246578,No,Basic,College,1/16/11,Employed,M,83302,Suburban,Married,67,24.0,21,0.0,1,Personal Auto,Personal L2,Offer1,Branch,345.6,Two-Door Car,Small,A
4785,LJ45155,Oregon,13992.20672,No,Basic,High School or Below,1/3/11,Employed,F,72687,Suburban,Married,75,25.0,30,0.0,2,Personal Auto,Personal L2,Offer1,Web,96.000717,Four-Door Car,Large,nan
4786,KA55650,Oregon,15146.79984,No,Extended,High School or Below,2/19/11,Unemployed,F,0,Suburban,Divorced,64,15.0,25,1.0,1,Corporate Auto,Corporate L3,Offer3,Branch,282.0,SUV,Medsize,A
4787,HY71976,Washington,2784.826476,No,Extended,High School or Below,2/24/11,Employed,F,68923,Urban,Married,64,5.0,68,0.0,2,Personal Auto,Personal L3,Offer1,Agent,453.6,Four-Door Car,Medsize,nan
4788,OE34471,Arizona,5563.267257,No,Premium,College,2/11/11,Employed,M,74153,Urban,Divorced,89,24.0,27,0.0,6,Personal Auto,Personal L3,Offer2,Branch,342.989674,SUV,Medsize,nan
4789,EK43950,California,3097.674466,Yes,Basic,High School or Below,1/27/11,Employed,M,59505,Suburban,Married,63,20.0,38,0.0,2,Personal Auto,Personal L3,Offer3,Agent,192.0,Two-Door Car,Medsize,A
4790,QT31934,California,10435.56957,No,Basic,Master,1/25/11,Medical Leave,F,11663,Urban,Married,66,22.0,34,0.0,2,Personal Auto,Personal L3,Offer1,Branch,656.872132,Luxury Car,Medsize,nan
4791,LY81726,Oregon,6190.256979,No,Extended,Bachelor,1/6/11,Unemployed,M,0,Suburban,Married,89,35.0,64,0.0,1,Personal Auto,Personal L1,Offer2,Call Center,716.4,Two-Door Car,Medsize,A
4792,JI85094,California,16248.33602,No,Basic,High School or Below,1/10/11,Unemployed,M,0,Suburban,Married,111,3.0,17,0.0,6,Personal Auto,Personal L2,Offer4,Agent,351.336039,Two-Door Car,Medsize,nan
4793,PL47486,California,2234.946206,No,Basic,College,1/7/11,Employed,F,44380,Urban,Married,72,12.0,36,0.0,2,Personal Auto,Personal L2,Offer3,Web,392.346212,Sports Car,Medsize,A
4794,KN91064,California,14422.07778,No,Premium,Bachelor,2/2/11,Unemployed,F,0,Suburban,Single,101,5.0,78,0.0,2,Personal Auto,Personal L3,Offer2,Agent,122.393347,Four-Door Car,Medsize,nan
4795,UB44543,California,15597.21956,No,Extended,Bachelor,2/15/11,Unemployed,F,0,Urban,Divorced,85,21.0,32,0.0,1,Personal Auto,Personal L2,Offer4,Agent,463.2,Sports Car,Small,A
4796,LH58015,California,5317.589497,No,Extended,College,2/13/11,Employed,F,57002,Suburban,Married,100,7.0,19,0.0,1,Personal Auto,Personal L3,Offer4,Branch,56.448187,SUV,Medsize,nan
4797,LW95297,California,22567.81592,No,Basic,High School or Below,2/17/11,Unemployed,M,0,Suburban,Single,107,19.0,26,0.0,1,Personal Auto,Personal L3,Offer2,Web,748.8,Two-Door Car,Medsize,A
4798,BK73550,California,12159.94186,No,Extended,Master,2/3/11,Employed,M,53707,Rural,Married,61,18.0,48,0.0,3,Personal Auto,Personal L3,Offer1,Branch,548.743138,Four-Door Car,Medsize,A
4799,ON55985,California,5313.334311,Yes,Basic,Bachelor,1/13/11,Unemployed,M,0,Suburban,Married,76,28.0,95,0.0,2,Personal Auto,Personal L2,Offer2,Call Center,477.6,SUV,Medsize,A
4800,KJ83542,California,2434.913539,Yes,Basic,College,1/17/11,Unemployed,F,0,Suburban,Single,115,14.0,75,0.0,1,Personal Auto,Personal L2,Offer2,Agent,296.201812,Four-Door Car,Medsize,nan
4801,QJ43218,California,28936.60472,Yes,Basic,High School or Below,1/9/11,Employed,F,0,Suburban,Married,119,13.0,65,1.0,2,Personal Auto,Personal L2,Offer2,Call Center,105.6,Sports Car,Medsize,A
4802,JE68921,California,6188.5731,No,Extended,Bachelor,2/11/11,Employed,F,48855,Suburban,Married,62,15.0,58,0.0,1,Personal Auto,Personal L3,Offer3,Web,573.883461,SUV,Medsize,nan
4803,RC34687,California,24441.36932,Yes,Extended,High School or Below,1/2/11,Employed,F,97355,Suburban,Single,129,19.0,16,0.0,3,Corporate Auto,Corporate L2,Offer2,Agent,424.8,Two-Door Car,Medsize,A
4804,DN50595,California,25335.93056,Yes,Basic,Master,1/2/11,Retired,F,22980,Suburban,Married,132,13.0,20,0.0,4,Corporate Auto,Corporate L3,Offer2,Branch,120.0,Four-Door Car,Medsize,A
4805,BX64938,California,5593.934676,Yes,Extended,High School or Below,2/6/11,Employed,F,55545,Suburban,Married,84,19.0,15,0.0,1,Personal Auto,Personal L3,Offer1,Branch,1056.0,Four-Door Car,Medsize,nan
4806,FM10576,California,4861.078515,No,Basic,Bachelor,2/22/11,Employed,M,61304,Suburban,Single,63,16.0,97,0.0,1,Personal Auto,Personal L1,Offer4,Branch,326.4,Two-Door Car,Medsize,A
4807,YO19126,California,7167.073756,Yes,Extended,Bachelor,1/9/11,Unemployed,F,0,Suburban,Single,93,28.0,5,0.0,1,Personal Auto,Personal L2,Offer2,Agent,518.4,Four-Door Car,Medsize,nan
4808,GZ62032,California,3653.424902,No,Extended,Master,2/25/11,Employed,M,57237,Urban,Married,87,20.0,92,1.0,1,Personal Auto,Personal L2,Offer1,Agent,497.6,Four-Door Car,Medsize,A
4809,JQ38644,California,4974.401943,No,Extended,College,2/15/11,Employed,M,58673,Suburban,Married,85,18.0,37,0.0,2,Personal Auto,Personal L2,Offer4,Agent,501.6,Four-Door Car,Medsize,A
4810,QH33628,California,3862.82878,No,Extended,Master,2/23/11,Employed,M,65753,Urban,Single,76,14.0,42,0.0,2,Personal Auto,Personal L1,Offer2,Agent,153.6,Four-Door Car,Medsize,nan
4811,RG25774,California,4858.499481,No,Basic,Bachelor,2/16/11,Employed,F,75920,Urban,Single,82,10.0,56,0.0,3,Personal Auto,Personal L3,Offer1,Call Center,552.0,Two-Door Car,Medsize,A
4812,YB65498,California,5483.5886,Yes,Basic,Master,2/12/11,Employed,F,53735,Urban,Married,98,14.0,57,0.0,2,Corporate Auto,Corporate L2,Offer3,Branch,552.0,Four-Door Car,Medsize,A
4813,RG35523,California,6026.260294,Yes,Basic,Bachelor,1/27/11,Employed,F,47429,Suburban,Single,93,13.0,62,0.0,1,Personal Auto,Personal L2,Offer2,Agent,634.4,Four-Door Car,Medsize,nan
4814,PG53960,California,5449.10617,No,Extended,Bachelor,2/21/11,Unemployed,F,0,Suburban,Divorced,64,20.0,92,0.0,1,Personal Auto,Personal L3,Offer2,Web,190.153681,Four-Door Car,Small,A
4815,YZ20291,California,4843.624818,Yes,Basic,Bachelor,2/26/11,Employed,F,50075,Suburban,Married,105,15.0,68,0.0,4,Personal Auto,Personal L3,Offer2,Branch,166.351484,Two-Door Car,Medsize,nan
4816,WO10426,California,6888.328401,Yes,Basic,High School or Below,1/1/11,Employed,F,82367,Suburban,Married,65,23.0,34,0.0,1,Personal Auto,Personal L3,Offer2,Agent,153.1,Two-Door Car,Medsize,A
4817,KO72255,California,10155.75511,No,Premium,High School or Below,2/10/11,Employed,M,79558,Suburban,Married,89,16.0,3,0.0,5,Personal Auto,Personal L2,Offer2,Branch,422.4,SUV,Medsize,A
4818,XN59796,California,3341.267286,Yes,Basic,Bachelor,2/23/11,Employed,F,24451,Urban,Married,78,13.0,66,0.0,1,Personal Auto,Personal L1,Offer2,Agent,408.0,Four-Door Car,Medsize,nan
4819,FP25289,California,9447.234825,No,Extended,Bachelor,1/28/11,Employed,M,64210,Urban,Married,91,1.0,69,0.0,2,Personal Auto,Personal L3,Offer1,Branch,156.754312,Four-Door Car,Medsize,A
4820,PE35317,California,5862.942572,No,Extended,High School or Below,2/9/11,Employed,F,64397,Suburban,Married,71,17.0,34,1.0,3,Personal Auto,Personal L3,Offer2,Agent,636.0,Four-Door Car,Small,nan
4821,IM49861,California,4019.026128,Yes,Basic,College,2/14/11,Employed,F,38059,Suburban,Married,90,28.0,58,0.0,3,Personal Auto,Personal L3,Offer2,Agent,896.4,Four-Door Car,Medsize,A
4822,YF58053,California,22174.78323,No,Extended,Master,1/7/11,Unemployed,F,0,Suburban,Single,111,3.0,18,0.0,4,Personal Auto,Personal L2,Offer1,Branch,674.4,Four-Door Car,Medsize,nan
4823,XC91005,California,5508.663661,Yes,Basic,College,1/16/11,Employed,F,60754,Suburban,Married,76,20.0,15,0.0,2,Personal Auto,Personal L3,Offer2,Branch,317.900353,SUV,Medsize,A
4824,BX11684,California,18698.27738,No,Extended,Master,1/11/11,Employed,M,57532,Urban,Married,65,35.0,97,0.0,2,Personal Auto,Personal L3,Offer2,Web,1056.0,Four-Door Car,Medsize,A
4825,YF48728,California,4291.846866,Yes,Basic,High School or Below,1/12/11,Employed,M,67010,Urban,Single,89,21.0,55,0.0,2,Personal Auto,Personal L2,Offer2,Agent,192.0,Four-Door Car,Medsize,nan
4826,WW49381,California,3621.812038,No,Premium,College,2/19/11,Unemployed,F,0,Suburban,Single,121,24.0,63,0.0,2,Personal Auto,Personal L3,Offer3,Agent,308.188682,Four-Door Car,Medsize,nan
4827,LJ46610,California,16792.99343,Yes,Extended,Master,1/24/11,Unemployed,F,0,Suburban,Married,67,2.0,41,0.0,2,Personal Auto,Personal L1,Offer2,Agent,105.6,Sports Car,Small,A
4828,WP73656,California,5561.560159,Yes,Extended,Bachelor,1/19/11,Employed,F,54642,Urban,Single,92,35.0,80,1.0,3,Personal Auto,Personal L3,Offer2,Branch,294.490143,Four-Door Car,Medsize,nan
4829,YG68479,California,8577.453001,Yes,Basic,High School or Below,2/2/11,Unemployed,M,0,Suburban,Married,110,26.0,72,0.0,2,Personal Auto,Personal L3,Offer3,Branch,113.073853,Four-Door Car,Medsize,A
4830,PL35129,California,6885.567262,No,Extended,College,1/4/11,Employed,M,75943,Urban,Married,67,14.0,89,0.0,1,Personal Auto,Personal L1,Offer1,Web,224.729733,Four-Door Car,Medsize,nan
4831,UI63950,California,7027.129743,No,Extended,College,1/11/11,Employed,F,58053,Suburban,Married,78,24.0,53,0.0,4,Personal Auto,Personal L3,Offer1,Agent,284.03799,Four-Door Car,Medsize,A
4832,RN85418,California,14897.86847,Yes,Extended,Master,2/22/11,Employed,F,61755,Suburban,Married,74,15.0,45,0.0,1,Personal Auto,Personal L1,Offer3,Web,388.8,Two-Door Car,Medsize,nan
4833,MT32059,California,7925.108999,No,Extended,Master,1/28/11,Employed,F,67636,Suburban,Married,66,6.0,66,0.0,1,Personal Auto,Personal L1,Offer1,Branch,960.0,Four-Door Car,Medsize,A
4834,VG35884,Oregon,3821.205298,No,Premium,Bachelor,2/5/11,Employed,M,80954,Rural,Married,100,2.0,23,0.0,2,Personal Auto,Personal L1,Offer4,Agent,636.0,Four-Door Car,Medsize,A
"""

# Process the data
df_first_quarter = process_insurance_data(data)

# Display the DataFrame
import ace_tools as tools; tools.display_dataframe_to_user(name="First Quarter Insurance Data", dataframe=df_first_quarter)
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the data
data = pd.read_csv('/mnt/data/insurance_data.csv')

# Clean the data
data_cleaned = data[['Monthly Premium Auto', 'Total Claim Amount']].dropna()

# Calculate correlation
correlation = data_cleaned.corr().iloc[0, 1]

# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='Monthly Premium Auto', y='Total Claim Amount', data=data_cleaned, scatter_kws={'s':10}, line_kws={'color':'red'})
plt.title(f'Relationship between Monthly Premium Auto and Total Claim Amount\nCorrelation: {correlation:.2f}')
plt.xlabel('Monthly Premium Auto')
plt.ylabel('Total Claim Amount')
plt.grid(True)
plt.show()

correlation
````
